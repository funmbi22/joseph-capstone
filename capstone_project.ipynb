```python
# customers table
customer_df = df[['customername']].reset_index(drop=True)
customer_df['customer_id'] = customer_df.index + 1
customer_df = customer_df[['customer_id', 'customername']]
```

```python
# product table
product_df = df[['category', 'sub-category', 'amount', 'quantity', 'profit']].reset_index(drop=True)
product_df['product_id'] = product_df.index + 1
product_df = product_df[['product_id', 'category', 'sub-category', 'amount', 'quantity', 'profit']]
```

```python
# orders table
orders_df = df[['order_id', 'paymentmode', 'order_date', 'state', 'city']].reset_index(drop=True)
orders_df['product_id'] = np.random.choice(product_df['product_id'], size=len(orders_df))
orders_df['customer_id'] = np.random.choice(customer_df['customer_id'], size=len(orders_df))
```
